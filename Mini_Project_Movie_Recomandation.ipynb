```python
# Import Modules

import pandas as pd
import numpy as np
import re
import nltk
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('IMDB_OMDB_Kaggle_TestSet_OMDB_Detailed.csv')
df.head()
```

```python
df['Plot'][4]
```

```python

##Data Preprocessing: Lowercase, remove punctuation, and normalize spaces
df['Clean_plot'] = df['Plot'].str.lower()
df['Clean_plot'] = df['Clean_plot'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))
df['Clean_plot'] = df['Clean_plot'].apply(lambda x: re.sub(r'\s+', ' ', x))
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
##Tokeniza the Sentence

import nltk
nltk.download('punkt', force=True)
nltk.download('punkt_tab', force=True)  # Add this line

from nltk.tokenize import word_tokenize

df['Clean_plot'] = df['Clean_plot'].apply(lambda x: word_tokenize(x))
df['Clean_plot']
```

```python
# Remove Stopwords

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')  # Only needed once

stop_words = stopwords.words('english')
plot = []

for sentence in df['Clean_plot']:
    temp = []
    for word in sentence:
        if word not in stop_words and len(word) >= 3:
            temp.append(word)
    plot.append(temp)

plot  # This will contain your cleaned, tokenized plots
```

```python
df['Clean_plot'] = plot
```

```python
df['Clean_plot']
```

```python
df.head()
```

```python
df['Genre'] = df['Genre'].apply(lambda x: [i.strip() for i in x] if isinstance(x, list) else [i.strip() for i in x.split(',')]) 
df['Actors'] = df['Actors'].apply(lambda x: [i.strip() for i in x[:4]] if isinstance(x, list) else [i.strip() for i in x.split(',')[:4]])
df['Director'] = df['Director'].apply(lambda x: [i.strip() for i in x] if isinstance(x, list) else [i.strip() for i in x.split(',')])
```

```python
df['Genre'][2]
```

```python
columns = ['Clean_plot', 'Genre', 'Actors', 'Director']
l = []
for i in range (len(df)):
    words = ''
    for col in columns:
        words += ' '.join(df[col][i]) + ' '
        l.append(words)
        l
```

```python
df['Clean_input'] = l[:len(df)]
df = df[['Title', 'Clean_input']]
df.head()
```

```python
## Features Extraction:
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df['Clean_input'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sin = cosine_similarity(features, features)
print(cosine_sin)
```

```python
# Movie Recomandation

index = pd.Series(df['Title'])
index.head()
```

```python
def recommend_movies(title, cosine_sin):
    movies = []
    
    # Get index of the movie
    idx = df[df['Title'] == title].index[0]
    print(f"Index of '{title}':", idx)
    
    # Get similarity scores and sort them
    score = pd.Series(cosine_sin[idx]).sort_values(ascending=False)
    top10 = list(score.iloc[1:11].index)  # Skip the first (same movie)
    print("Top 10 similar movie indices:", top10)
    
    # Fetch movie titles
    for i in top10:
        movies.append(df['Title'][i])
    
    return movies
```

```python
recommend_movies( "Pirates of the Caribbean: At World's End", cosine_sin)
```
